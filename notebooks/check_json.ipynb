# Check JSON Structure

Quick notebook to inspect JSON files (arrays of dicts).

```python
import json

# Set your JSON file path here
JSON_PATH = '../src/training/rest/outputs/filtered/filtered.json'
```

```python
# Load JSON
with open(JSON_PATH, 'r') as f:
    data = json.load(f)

print(f"Type: {type(data)}")
print(f"Length: {len(data)}")
```

```python
# Check first element
print("First element:")
print(json.dumps(data[2], indent=2))
```

```python
# Keys in first element
print("Keys:", list(data[0].keys()))
```
